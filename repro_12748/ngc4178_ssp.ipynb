```python
import numpy as np
import matplotlib.pyplot as plt
import fsps
```

```python
sp = fsps.StellarPopulation(zcontinuous=1, dust_type=2, dust2=0.2)
tages = np.linspace(0.002, 0.02, 100)
spec_arr = np.zeros_like(tages)
wave, spec = sp.get_spectrum(tage=0.004, peraa=True)   #Just an example
print(wave[5270], wave[5282], wave[5291]) # These correspond to [Ne II], [Ne V] and [Ne III]
print(spec[5270], wave[5282], wave[5291])
spec_arr = np.zeros((len(tages), len(wave)))
```

```python
for i, age in enumerate(tages):
    wave, spec_arr[i] = sp.get_spectrum(tage=age, peraa=True)
    if not wave[5270] == 128200.00:
        print('Wavelength bins are changing at ' + age)
```

```python
%matplotlib inline
def plotfig(title, xlabel, ylabel, xdata, ydata):
    plt.figure(figsize=(4, 3))
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(xdata, ydata)
    plt.show()
```

```python
plotfig('[NeV] luminosity with age', 'Time (in Myr)', '[Ne V] luminosity', tages*1000, spec_arr[:, 5282]*0.5*10**4)
plotfig('[Ne V]/[Ne III] ratio with age', 'Time (in Myr)', '[Ne V]/[Ne III]', tages*1000, spec_arr[:, 5282]/spec_arr[:, 5291])
plotfig('[Ne V]/[Ne II] ratio with age', 'Time (in Myr)', '[Ne V]/[Ne II]', tages*1000, spec_arr[:, 5282]/spec_arr[:, 5270])
```
